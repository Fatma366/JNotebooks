```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
counties = ['Kilifi', 'Kwale', 'Lamu']
prevalence = [11.6, 11.3, 9.5]
severity = [70.6, 60.5, 56.7]
```

```python
# Step 3: Set up the figure and axes with increased figsize
fig, ax = plt.subplots(figsize=(9, 7))

# Step 4: Create the bar plot for prevalence
x_pos = np.arange(len(counties))
bar_width = 0.45

# Specify custom colors for the bars
prevalence_colors = ['red', 'green', 'blue']
severity_colors = ['orange', 'purple', 'yellow']

ax.bar(x_pos - bar_width/2, prevalence, bar_width, label='Prevalence (%)', edgecolor='black')

# Step 5: Create the bar plot for severity
ax.bar(x_pos + bar_width/2, severity, bar_width, label='Severity (%)', edgecolor='black')


# Step 6: Set the axis labels, title, and legend
ax.set_xlabel('Counties')
ax.set_ylabel('Disease Prevalence and Severity (%)')
ax.set_title('Prevalence and Severity scores for LY symptomatic plants')
ax.set_xticks(x_pos)
ax.set_xticklabels(counties)
ax.legend()

# Set the y-axis limits
ax.set_ylim(0, 90)

#Step 7: Add percentage labels on top of each bar
for i, v in enumerate(prevalence):
    ax.text(i - bar_width/2, v + 0.5, str(v), color='black', ha='center')

for i, v in enumerate(severity):
    ax.text(i + bar_width/2, v + 0.5, str(v), color='black', ha='center')

# Display the plot
plt.show()

```
