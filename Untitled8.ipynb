```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the font family and font size
sns.set(font='Times New Roman', font_scale=1.0)

# Read data from Excel file (replace 'data.xlsx' with your file path)
df = pd.read_excel("C:/Users/Admin/Desktop/RESULTS 2.0/Colony Moph, Gram stain, 101 - coded.xlsx", engine='openpyxl')

# Set the 'Isolate' column as the index
df.set_index('Sample ID', inplace=True)

# Create the heatmap with dendrograms using seaborn's clustermap
plt.figure(figsize=(8, 6))
clustermap = sns.clustermap(df, annot=False, cmap='coolwarm', cbar_kws={"shrink": 0.5})
ax = clustermap.ax_heatmap
ax.set_xlabel('Morphological Descriptors', fontsize=14, fontweight='bold')
ax.set_ylabel('Bacterial Isolates', fontsize=14, fontweight='bold')

# Darken the tick labels on the x-axis and y-axis
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, fontweight='bold')
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12, fontweight='bold')

plt.show()
```
